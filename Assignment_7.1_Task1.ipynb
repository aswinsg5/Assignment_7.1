```python
adult_dataset=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None)
```

```python
col_names=pd.Series(data=("age","workclass","fnlwgt","education","education_num","marital_status","occupation","relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","salary_class"))
```

```python
adult_dataset=pd.DataFrame(adult_dataset)
adult_dataset.columns=col_names
```

<b>2. Show me the average hours per week of all men who are working in private sector</b>

```python
sqldf("SELECT avg(hours_per_week) FROM sqladb where workclass = ' Private' ;")
```

<b>3. Show me the frequency table for education, occupation and relationship, separately</b>

```python
#Education Frequency
sqldf(" SELECT education,count(education)  from sqladb group by education")
```

```python
# Occupation Frequency
sqldf(" SELECT occupation,count(occupation)  from sqladb group by occupation")
```

```python
# Relationship Frequency
sqldf(" SELECT relationship,count(relationship)  from sqladb group by relationship")
```

<b>4. Are there any people who are married, working in private sector and having a masters degree</b>

```python
sqldf("SELECT * from sqladb where marital_status in (' Married-civ-spouse', ' Separated',' Married-spouse-absent', ' Married-AF-spouse')  and education=' Masters' and workclass=' Private' ")
```

<b>5. What is the average, minimum and maximum age group for people working in different sectors</b>

```python
sqldf("SELECT workclass,avg(AGE),max(AGE),min(AGE) from sqladb group by workclass; ")
```

<b>6. Calculate age distribution by country</b>

```python
sqldf("SELECT native_country, avg(age) from sqladb group by native_country")
```

<b>7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'</b>

```python
sqldf("SELECT capital_gain,capital_loss,capital_gain - capital_loss as 'Net-Capital-Gain' from sqladb")
```
